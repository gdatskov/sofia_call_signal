### Now, First, expand nested columns

#### Copy/paste the notorious json expanding function lurking in the www:

```python
def json_to_series(text):
    keys, values = zip(*[(dct['label'], dct['value']) for dct in json.loads(text)])
    return pd.Series(values, index=keys)
```

#### And ask ChatGPT how to use it on our data:

```python
# First, open the file and json-convert it to python dict object:
def json_to_python_dict(file_path):
    data = []
    with open(file_path, encoding='utf-8') as file: # Open file and save JSON in list
        for line in file:
            obj = json.loads(line)
            data.append(obj)
    json_string = json.dumps(data)  # Convert data list to JSON string
    json_dict = json.loads(json_string)  # Load JSON string as a Python object
    return json_dict
    
```

#### Then convert these python dicts into pandas dataframes:

```python
# History DF:
history_dict = json_to_python_dict(history_file_path)
history_combined_dict = {
    'number': [obj['number'] for obj in history_dict],
    'history': [obj['history'] for obj in history_dict]
}
history_df = pd.concat([pd.DataFrame(history_combined_dict), pd.DataFrame(list(history_combined_dict['history']))], axis=1).drop('history', 1)

# Answers DF:
answers_dict = json_to_python_dict(answers_file_path)
answers_combined_dict = {
    'number': [obj['number'] for obj in answers_dict],
    'answers': [obj['answers'] for obj in answers_dict]
}
answers_df = pd.concat([pd.DataFrame(answers_combined_dict), pd.DataFrame(list(answers_combined_dict['answers']))], axis=1).drop('answers', 1)
```

#### And now actually implement function to expand the JSONs (The data format we have)

```python
def expand_json_df(df, column_to_keep='number', suffix_sep='_'):
    df_expanded = df[column_to_keep].to_frame()  # Create a dataframe with the 'number' column
    for col_name in df.columns:
        if col_name != column_to_keep:
            expanded_json = pd.json_normalize(df[col_name]).add_suffix(f"{suffix_sep}{col_name}")
            df_expanded = pd.concat([df_expanded, expanded_json], axis=1)
    return df_expanded
```

```python
# Expand the JSONs:
history_expanded = expand_json_df(history_df)
answers_expanded = expand_json_df(answers_df)
   
```

```python
def generate_square_area_df(latitude, longitude, distance_meters, dataframe):
    # Approximate conversion from meters to degrees at the equator
    meters_per_degree = 111000
    distance_degrees = distance_meters / meters_per_degree / 2
    
    lat_start = latitude - distance_degrees
    lat_end = latitude + distance_degrees
    long_start = longitude - distance_degrees
    long_end = longitude + distance_degrees
    
    # square_geolocation = ((lat_start, lat_end), (long_start, long_end))
    
    # Filter the dataframe based on the square geolocation
    mask = (
        (dataframe['latitude'] >= lat_start) &
        (dataframe['latitude'] <= lat_end) &
        (dataframe['longitude'] >= long_start) &
        (dataframe['longitude'] <= long_end)
    )
    
    return dataframe[mask]
```
