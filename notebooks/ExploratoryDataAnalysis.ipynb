### First, expand nested columns

```python
# Copy/paste the notorious json expanding function lurking in the www:
def json_to_series(text):
    keys, values = zip(*[(dct['label'], dct['value']) for dct in json.loads(text)])
    return pd.Series(values, index=keys)
```

```python
# And ask ChatGPT how to use it on our data:
# First, open the file and json-convert it to python dict object:
def json_to_python_dict(file_path):
    data = []
    with open(file_path, encoding='utf-8') as file: # Open file and save JSON in list
        for line in file:
            obj = json.loads(line)
            data.append(obj)
    json_string = json.dumps(data)  # Convert data list to JSON string
    json_dict = json.loads(json_string)  # Load JSON string as a Python object
    return json_dict
    
```

```python
# Then convert these python dicts into pandas dataframes:

# History DF:
history_dict = json_to_python_dict(history_file_path)
history_combined_dict = {
    'number': [obj['number'] for obj in history_dict],
    'history': [obj['history'] for obj in history_dict]
}
history_df = pd.concat([pd.DataFrame(history_combined_dict), pd.DataFrame(list(history_combined_dict['history']))], axis=1).drop('history', 1)

# Answers DF:
answers_dict = json_to_python_dict(answers_file_path)
answers_combined_dict = {
    'number': [obj['number'] for obj in answers_dict],
    'answers': [obj['answers'] for obj in answers_dict]
}
answers_df = pd.concat([pd.DataFrame(answers_combined_dict), pd.DataFrame(list(answers_combined_dict['answers']))], axis=1).drop('answers', 1)
```

```python
# And now function to expand the JSONs
def expand_json_df(df, column_to_keep='number', suffix_sep='_'):
    df_expanded = df[column_to_keep].to_frame()  # Create a dataframe with the 'number' column
    for col_name in df.columns:
        if col_name != column_to_keep:
            expanded_json = pd.json_normalize(df[col_name]).add_suffix(f"{suffix_sep}{col_name}")
            df_expanded = pd.concat([df_expanded, expanded_json], axis=1)
    return df_expanded
```

```python
# Expand the JSONs:
history_expanded = expand_json_df(history_df)
answers_expanded = expand_json_df(answers_df)
   
```

### Then, merge everything and work our way out

```python
# Unfortunately our scraped data for the signal category and entry code is crap (didnt scrape number along with the other data). 
# Fortunately the number is in the description, so it has to be extracted from there.
```

```python
# Extracting the number using regex and creating a new column
signal['number'] = signal['signal'].apply(lambda x: int(re.search(r'№(\d+)', x).group(1)) if re.search(r'№(\d+)', x) else None)
# Reordering the columns
signal = signal[['number', 'signal', 'category', 'sub_category', 'extraction_time']]
```

```python
# In order not to lose any data, outer join should be performed.
# Before merging, it would be nice to get rid of duplicates for future analysis of individual DFs.
```

```python
# Check for duplicates
duplicates = address.duplicated()

# Display the duplicate rows
duplicate_rows = address[duplicates]
```

```python
# Add DFs here (as strings):
list_of_df = ['address', 'answers_expanded', 'comments', 'data', 'history_expanded', 'signal', 'status']
```

```python
for df in list_of_df:
    globals()[df] = eval(df).drop_duplicates()
```

```python
# Determine the DF with most records (not that it matters when outer join is performed):
len_of_df = [len(eval(x)) for x in list_of_df]
dict_df_len = dict(zip(len_of_df, list_of_df))
df_most_records = dict_df_len[max(dict_df_len)]
print(f'The DF with most records is: >>> {df_most_records} <<<')
```

```python
# Some quick checks:
inner_join = pd.merge(signal, data, on='number', how='inner')
check_left = pd.merge(signal, data, on='number', how='left')
check_right = pd.merge(signal, data, on='number', how='right')
```

```python
# Left join has more records than right join:
len(check_left) > len(check_right)
```

```python
# Inner join has more records than right join:
len(inner_join) > len(check_right)
```

```python
# Inner join has more records than right join:
len(inner_join) < len(check_right)
```

```python
# NOW LETS JOIN!
join_list = [x for x in list_of_df if x != df_most_records]
full_df = eval(df_most_records)
for df in join_list:
    df = eval(f"{df}")
    
    current_columns = full_df.columns
    new_columns = df.columns
    
    df = df.drop(columns=[col for col in new_columns if col in current_columns and col != 'number'])
    full_df = full_df.merge(df, on='number', how='outer')
```
