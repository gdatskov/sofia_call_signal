```python
import json
import numpy
import pandas as pd
import io
```

```python
file_path = '../available_data/history.json'
```

```python
df = pd.read_json(file_path, lines=True)
```

```python
df
```

```python
pd.json_normalize(df.history[0])
```

```python
def json_to_series(text):
    keys, values = zip(*[(dct['label'], dct['value']) for dct in json.loads(text)])
    return pd.Series(values, index=keys)
```

```python
pd.concat([df, df['history'].apply(json_to_series)], axis=1)
```

```python
data = []

with open(file_path, encoding='utf-8') as file:
    for line in file:
        obj = json.loads(line)
        data.append(obj)
```

```python
json_string = json.dumps(data)  # Convert data list to JSON string
json_dict = json.loads(json_string)  # Load JSON string as a Python object
```

```python
json_dict = json.load(io.StringIO(data))
```

```python
pd.concat([pd.DataFrame(json_dict), pd.DataFrame(list(json_dict['answers']))], axis=1).drop('answers', 1)
```

```python
combined_dict = {
    'number': [obj['number'] for obj in json_dict],
    'history': [obj['history'] for obj in json_dict]
}

df = pd.concat([pd.DataFrame(combined_dict), pd.DataFrame(list(combined_dict['history']))], axis=1).drop('history', 1)
```

```python
df[df[2].notnull()]
```

```python
df_expanded = df['number'].to_frame()  # Create a dataframe with the 'number' column
for col_name in df.columns:
    if col_name != 'number':
        expanded_df = pd.json_normalize(df[col_name]).add_suffix(f"_{col_name}")
        df_expanded = pd.concat([df_expanded, expanded_df], axis=1)
```

```python
df_expanded
```

```python
expanded_columns = []  # Store the expanded columns

for col_name in [0, 1, 2]:
    expanded_df = pd.json_normalize(df[col_name]).add_suffix(f"_{col_name}")
    expanded_columns.append(expanded_df)

df_expanded = pd.concat([df['number']] + expanded_columns, axis=1)
```
